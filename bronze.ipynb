```python
if spark.catalog.tableExists("datamodelling.default.source_data"):
    last_load_date = spark.sql("select max(order_date) from datamodelling.default.source_data").collect()[0][0]
else:
    last_load_date = '1000-01-01'
```

```python
last_load_date
```

```python
# Set a very old date for the initial load, or query the max date for incremental loads


# Extract only new incremental data using PySpark
bronze_df = spark.sql(f"""
    SELECT * 
    FROM datamodelling.default.source_data 
    WHERE order_date > '{last_load_date}'
""")
```

```python
# Create a temporary view to use in standard SQL
bronze_df.createOrReplaceTempView("bronze_source")
```

```python
%sql
CREATE OR REPLACE TABLE datamodelling.bronze.bronze_table AS
SELECT * FROM bronze_source;
```
